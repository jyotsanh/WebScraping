```python
def start(data):
    df = dict()
    name = []
    gpu = []
    ram = []
    sto = []
    display = []
    for i in range(len(data)):
        dataa = data[i]
        if(dataa=='MSI Modern 14 B4MW Ryzen 7 4700U / 8GB RAM / 256GB SSD / 14" FHD Display'):
            model = dataa.split('/')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        elif(dataa=='Lenovo Yoga Slim 7 15ITL05 i5 11th Gen / 16GB RAM / 512GB SSD / 15.6" FHD display'):
            model = dataa.split('/')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        elif(dataa=='Lenovo ThinkPad L14 Gen 2 Ryzen 5 Pro 5650 / 16GB RAM / 512GB SSD / 14" FHD Display'):
            model = dataa.split('/')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        elif(dataa=='Lenovo Yoga Slim 7 Pro 14ARH7 Ryzen 7 6800HS / 16GB RAM / 1TB SSD / 14" 2.8K (2880x1800)  display'):
            model = dataa.split('/')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        elif(dataa=="Dell Inspiron 7306 2-in-1 i7 11Th Gen / 16GB RAM / 512GB SSD / 13.3'' FHD 360-degree Touch Screen Display"):
            model = dataa.split('/')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        elif(dataa=='Apple M1 Pro MacBook Pro 14-inch M1 Pro Chip, 16GB RAM, 512GB Storage, 14.2-inch 120Hz Liquid Retina XDR Display'):
            model = dataa.split(',')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        elif(dataa=='Apple M1 Pro MacBook Pro 14-inch M1 Pro Chip, 16GB RAM, 1TB Storage, 14.2-inch 120Hz Liquid Retina XDR Display'):
            model = dataa.split(',')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3]) 
        elif(dataa=='Apple M1 Pro MacBook Pro 16-inch M1 Pro Chip, 16GB RAM, 512GB/1TB Storage, 16.2-inch 120Hz Liquid Retina XDR Display'):
            model = dataa.split(',')
            name.append(model[0])
            gpu.append('Nan')
            ram.append(model[1])
            sto.append(model[2])
            display.append(model[3])
        else:
           
            model = dataa.split('/')
            
            name.append(model[0])
            gpu.append(model[1])
            ram.append(model[2])
            sto.append(model[3])
            display.append(model[4])
        
        
    df = { 'model':name,'gpu':gpu,'ram':ram,'sto':sto }
    final = pd.DataFrame(df)
    return final 
        
```

```python
final.rename(columns={'model':'Model','gpu':'GPU','ram':'RAM','sto':'Storage'},inplace=True)
```

```python
final['GPU']=final['GPU'].apply(lambda x: x.replace('8GB RAM','GTX 1650TI'))
final['GPU']=final['GPU'].apply(lambda x: x.replace('16GB RAM','RTX 3050Ti'))
final['GPU']=final['GPU'].apply(lambda x: x.replace('32GB RAM','RTX 3070Ti'))
final['Storage']=final['Storage'].apply(lambda x: x.replace('GTX 1650TI','512GB SSD'))
final['Storage']=final['Storage'].apply(lambda x: x.replace('RTX 3050Ti','1TB HDD'))
```
